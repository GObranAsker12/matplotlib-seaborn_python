```python
import numpy as np
# importing matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
f=[2,6,5,7,9]
sns.kdeplot(f)
plt.show()
```

```python
f=[2,6,5,7,9]
plt.plot(f)
plt.show()
```

```python
x=[8,9,63,6,2,7]
y=[6,8,9,2,19,30]
plt.plot(x,y)
plt.xlabel('x Number')
plt.ylabel('y Number' )
plt.show()
```

```python
x=np.arange(0,10,0.1)
a=np.sin(x)
b=np.cos(x)
c=np.exp(x/10)

plt.plot(x,a,'b-',label='sin')
plt.plot(x,b,'r--',label='cos')
plt.plot(x,c,'g-',label='+x')
plt.legend(loc='upperleft')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.show()

```

```python
def p1(x): return 4**x-4*x**2+3*2
def p2(x):return np.sin(10*x)*10

x=np.linspace(-3,3,2000)
plt.plot(x,p1(x),x,p2(x))
plt.show
```
